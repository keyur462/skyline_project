```python
def sortFilesById(files):
    ids = []
    for f in files:
        ids.append(int(f[f.find('_')+1:f.find('.txt')]))
    ids, files = zip(*sorted(zip(ids, files)))
    return files
```

```python
def createWeightedJaccardIndex(data, M=16, efC=200, threads=4, post=1):
    index = nmslib.init(method='hnsw', space='WeightedJaccard')
    index.addDataPointBatch(data)

    params = {
        'M': M,
        'efConstruction': efC,
        'indexThreadQty': threads,
        'post': post
    }

    t0 = time.time()
    index.createIndex(params)
    t1 = time.time()

    print("Index built in {:.2f} sec".format(t1 - t0))
    return index
```
